```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("GWQ_SDP.csv")

data['SDP'] = pd.to_numeric(data['SDP'], errors='coerce')
data['Calcium'] = pd.to_numeric(data['Calcium'], errors='coerce')

# Drop rows with NaN values
data.dropna(subset=['SDP', 'Calcium'], inplace=True)

beta0 = 53.8226
beta1 = 0.00004718

data['y_hat'] = beta0 + beta1 * data['SDP']

data['residuals'] = data['Calcium'] - data['y_hat']

# Plot groundwater quality indicator against SDP
plt.scatter(data['SDP'], data['Calcium'])
plt.xlabel('SDP')
plt.ylabel('Calcium')
plt.title('GWQ Indicator vs. SDP')
plt.show()

# Plot residuals against SDP
plt.scatter(data['SDP'], data['residuals'])
plt.xlabel('SDP')
plt.ylabel('Residuals')
plt.title('Residuals vs. SDP')
plt.show()
```

```python
'''
The first plot, with groundwater quality (Calcium) on the Y-axis and SDP on the X-axis, shows the relationship between these two variables. This plot is what we
would expect, as it provides a visualization of how groundwater quality varies with changes in the socioeconomic development parameter (SDP). Typically, one would 
anticipate observing some sort of trend or pattern in this plot, as changes in SDP could potentially influence groundwater quality.

The second plot, with residuals on the Y-axis and SDP on the X-axis, helps to assess the adequacy of the regression model. Ideally, the residuals should be 
randomly scattered around zero without any noticeable patterns or trends. However, if we observe a pattern or trend in the residuals, it indicates that the model 
might not adequately capture the relationship between groundwater quality and SDP. This could suggest potential issues such as omitted variables, nonlinear 
relationships, or violations of model assumptions.

In summary, while the first plot represents the actual relationship between groundwater quality and SDP, the second plot aids in evaluating the performance of 
the regression model by examining the residuals. Both plots are essential in understanding and validating the relationship between the variables and the 
regression model.
'''
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import math

data = pd.read_csv("GWQ_SDP.csv")

data['SDP'] = pd.to_numeric(data['SDP'], errors='coerce')
data['Calcium'] = pd.to_numeric(data['Calcium'], errors='coerce')

data.dropna(subset=['SDP', 'Calcium'], inplace=True)

beta0 = 53.8226
beta1 = 0.00004718

data['y_hat'] = beta0 + beta1 * data['SDP']

data['residuals'] = data['Calcium'] - data['y_hat']

# Plot histogram of residuals
plt.hist(data['residuals'], bins=20)
plt.xlabel('Residuals')
plt.ylabel('Frequency')
plt.title('Histogram of Residuals')
plt.show()

# Verify sum of residuals
residual_sum = data['residuals'].sum()
print("Sum of residuals:", math.floor(residual_sum))
```
